##Hypothesis Testing 1

##Hypothesis testing of two independent data set created out of the original data.
###1.A data set consist of movies with action gener.
###2.The second set of data set consists of movies with other gener other than action movies

####Claim: The gross income of action movies is the heighest
####Alternative : Other movies gross income is < action movies gross income
####statistic test= Z test
####alpha value is considerend to be 0.05
####I will be using the left tailed test

```python
#will be seperting the data into two different data frames to find the mean of each data set
Other_df = Movies_df[Movies_df['genre']!= 'Action']
Action_df = Movies_df[Movies_df['genre'] =='Action']
print('The mean profit for all genre other than Action = {}'.format(np.mean(Other_df['gross'])))
print('The mean profit of Action genre = {}'.format(np.mean(Action_df['gross'])))
```

```python
#Visualizing the seperated data set to check the normality.
#Although according to central limit theorem n>30 is considered to be normally distributed Iam verifying it in the vsualization form.
import warnings
warnings.filterwarnings("ignore") #to ignore the futurewarning of distplot
sn.distplot(Other_df['gross']) #for getting the distribution plot for 'profit'from genres other than action
plt.show()
sn.distplot(Action_df['gross'])#for getting the distribution plot for 'profit' from Action genre
plt.show()
```

```python
# To find standard deviation
import math as math
import statistics as stats
s1=stats.stdev(Action_df.gross)
print("The std of action movies data set",s1)
s1_sqr=math.pow(s1,2)
print("The square of s1 is", s1_sqr)
s2=stats.stdev(Other_df.gross)
print("The std of other movies data set", s2)
s2_sqr=math.pow(s2,2)
print("The square of s2 is", s2_sqr)
n1=1705
n2=5963
std=math.sqrt((s1_sqr/n1) + (s2_sqr/n2))
print("The standard deviation of action movies and other movie sample are" , std)
```

```python
import scipy as scp
alpha=0.05
mean2=stats.mean(Action_df.gross)
mean1=stats.mean(Other_df.gross)
#Using the z test
# Calculate the z-score for the difference in means
z_value = (mean2 - mean1) / np.sqrt((s1**2 / n1) + (s2**2 / n2))
print("The z  score is",z_value)

#alpha value is
print("The alpha value is", alpha)

if z_value > alpha:
    print("Fail to reject the null hypothesis - we have nothing else to say")
else:
    print("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

###With this we can infer that the action movies may have the heighest gross income

##Hypothesis testing 2

####Claim = The maximum votes obtaine = 2400000
####Alternate = The votes obtained is != 2400000
####The total available is grater than 30 so I will use z-test
####I will use two tail test for this hypothesis testing

```python
#Iam writing a function to rndom sample creating a sample out of population
# function that creates random sample
def random_sampling(df, n):
    random_sample = np.random.choice(df,replace = False, size = n)
    return(random_sample)
```

```python
N=1500
randomSample = random_sampling(Movies_df.votes, N)
randomSample
```

```python
#To find the sample mean
sample_mean = stats.mean(randomSample)
print("The sample mean is ", sample_mean)
```

```python
differences = [randomSample - sample_mean for i in range(len(randomSample))]
print(differences)

Differences_square = [(randomSample - sample_mean)**2 for i in range(len(randomSample))]
print(Differences_square)
```

```python
Difference_sum=np.sum(differences)
print("The summation value is ", Difference_sum)

Difference_sqr_sum=np.sum(Differences_square)
print("The summation value is ", Difference_sqr_sum)
```

```python
# standard deviation
n_new=1500
Differences_sum = -1.7626909539103508e-05
Differences_sqr_sum = 5.650655689194872e+16

standard_deviation = (((Differences_sum - Differences_sqr_sum)**2 / (n_new - 1)) ** 0.5)
print("standard_deviation:", standard_deviation)
```

```python
#To calculate the Z-score
n_new=1500
sample_mean= sample_mean


print("The sample mean is",sample_mean)
population_mean = stats.mean(Movies_df.votes)
print("The population mean is",population_mean)
std=standard_deviation
print("The standard deviation of the sample is",std)

Zscore=(sample_mean-population_mean)/(std/np.sqrt(n_new))
print("The zscore value is" , Zscore)
```

```python
#find Z critical value
import scipy
from scipy.stats import norm
sample_alpha=0.05/2
Z_critical=scipy.stats.norm.ppf(sample_alpha)
print("The Z critical value is ",Z_critical)

#find p-value
p_value = scipy.stats.norm.sf(abs(Zscore))

print('p_value: %0.5f' % (p_value))

if p_value > alpha:
    print("Fail to reject the null hypothesis - we have nothing else to say")
else:
    print("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

##In this case the random sampling of 1500 has rejected the null hypothesis  With this I can conclude the sample representents the population. This samplining can be conducted a few more times to verify the hypothesis further.

```python
#To find the samll sample deviation the steps are
small_differences = [randomSample1 - smal_sample_mean for i in range(len(randomSample1))]
print(small_differences)

small_differences_square = [(randomSample1 - smal_sample_mean)**2 for i in range(len(randomSample1))]
print(small_differences_square)
```

```python
small_Difference_sum=np.sum(small_differences)
print("The summation value is ", small_Difference_sum)

small_Difference_sqr_sum=np.sum(small_differences_square)
print("The summation value is ", small_Difference_sqr_sum)
```

```python
# standard deviation
small_new=25
small_Differences_sum = small_Difference_sum
small_Differences_sqr_sum = small_Difference_sqr_sum

small_standard_deviation = (((small_Differences_sum - small_Differences_sqr_sum)**2 / (small_new - 1)) ** 0.5)
print("standard_deviation:", small_standard_deviation)
```

```python
#To find the t-score
small_alpha=0.01
small_dash=small_Difference_sum/small_new
small_t= (small_dash-0)/(small_standard_deviation/stats.sqrt(small_new))
print("The t score is",small_t)

if small_t>small_alpha:
  print("Fail to reject the null hypothesis ")
else:
    print("Reject the null hypothesis - suggest the alternative hypothesis is true")

```

```python
#Increasing the alpha value to check for variation
small_alpha=0.05
small_dash=small_Difference_sum/small_new
small_t= (small_dash-0)/(small_standard_deviation/stats.sqrt(small_new))
print("The t score is",small_t)

if small_t>small_alpha:
  print("Fail to reject the null hypothesis ")
else:
    print("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

```python
#Regression model with a seperate dataframe with 5 dependent and 1 dependent variable which is multi regression
import statsmodels.api as smf
from patsy import dmatrices
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

Reg_df = Movies_df[['gross','budget', 'score', 'votes', 'year', 'runtime']]
Reg_df = smf.add_constant(Reg_df, prepend=True)
Reg_df
```

```python
#Splitting the train and test data
# Select the gross column as the feature to be predicted (y)
y = Reg_df['gross']

# Remove the gross column to create the training data
X = Reg_df.drop('gross', axis=1)

# Perform a 75% train and 25% test data split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=1)

print("{} rows in test set vs. {} in training set. {} Features.".format(X_test.shape[0], X_train.shape[0], X_test.shape[1]))
```

```python
#Printing all the shape of the train and test values to check for the array sizes to plot them in future
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predict
y_pred = regressor.predict(X_test)

#Accuracy
from sklearn.metrics import r2_score
r2_test = r2_score(y_test, y_pred)

from sklearn.metrics import mean_squared_error
from math import sqrt
mse_test = sqrt(mean_squared_error(y_test, y_pred))
```

```python
#Printing the r2 error and mean square error of the test value
print("The r2 error of the test model is",r2_test)
print("The mean square error of the test model is",mse_test)
```
